```python
#get dailiy quotes and then each daily close value for company code 86970
r = requests.get("https://api.jpx-jquants.com/v1/prices/daily_quotes?code=86970", headers=headers).json()['daily_quotes']
df = pd.DataFrame(r)
df = df[['Date','Close']]
df.set_index(df['Date'], inplace=True)
df=df.drop(['Date'], axis=1)
df.sort_index()
df=df.tail(180)
#df.loc[pd.to_datetime(df['Date'], dayfirst=True) > pd.to_datetime(df['Date']]
```

```python
df.plot()
```

```python
plt.style.use('fivethirtyeight')
plt.ylabel("Close")
plt.plot(df.index,df['Close'].values,linewidth=0.5)
plt.legend()
```

```python
# 10 days and 50 days moving average
df["20_MA"] = df['Close'].rolling(20).mean()
df["50_MA"] = df['Close'].rolling(50).mean()
df
```

```python
df.dropna(inplace=True)
df
```

```python
#SHORT or LONG 
df["SHORT_OR_LONG"] = np.where(df["20_MA"] > df["50_MA"], 1,0)
df
```

```python
df['Signal'] = df["SHORT_OR_LONG"].diff()
df
```

```python
df.dropna(inplace=True)
df
```

```python
plt.figure(figsize=(20,10))
df["Close"].plot(label="Close", color="Black")
df["20_MA"].plot(label="20_MA", color="Blue")
df["50_MA"].plot(label="50_MA", color="Green")
#Add buy signal 
plt.plot(df[df["Signal"] ==1].index, df["20_MA"][df["Signal"] ==1],"^", markersize=12, color="Green", label="Buy") 
#Add sell signal 
plt.plot(df[df["Signal"] ==-1].index, df["20_MA"][df["Signal"] ==-1],"^", markersize=12, color="Red", label="Sell") 

plt.legend()
```
